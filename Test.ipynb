```python
from xlrd import open_workbook
import matplotlib.pyplot as plt 
import requests 
import pandas as pd
import numpy as np
 
```

```python
test = pd.read_excel('Test_File_NP.xlsx')
test.head()
```

```python
winners = ["THE KING'S SPEECH", "THE ARTIST", "ARGO", "12 YEARS A SLAVE" , "BIRDMAN" , "SPOTLIGHT" , "MOONLIGHT" , "THE SHAPE OF WATER"]

winning_movies = test.loc[test["Title"].isin(winners), :]
winning_movies["Profit"] = winning_movies["Worldwide Gross"] - winning_movies["Budget"]
winning_movies['Earning per $ spent'] = winning_movies["Worldwide Gross"]/winning_movies["Budget"]
winning_movies['Earning per min'] = winning_movies["Worldwide Gross"]/winning_movies["Run Time"]
winning_movies.head()
```

```python
Earning_per_dollar = winning_movies['Earning per $ spent']
x_axis = winning_movies["Year"]
plt.plot(x_axis, Earning_per_dollar, color="b")

plt.show()
```

```python
Earning_per_min = winning_movies['Earning per min']
x_axis = winning_movies["Year"]
plt.scatter (x_axis, Earning_per_min,marker="o", facecolors="r", edgecolors="b")
plt.show()
```
