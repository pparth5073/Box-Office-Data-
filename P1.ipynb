```python
import requests
import json
import pandas as pd
import time
from pprint import pprint
```

```python
# Note that the ?t= is a query param for the t-itle of the
# movie we want to search for.
#https://api.themoviedb.org/3/search/movie?api_key={api_key}&query=Jack+Reacher
       
url_ID = "https://api.themoviedb.org/3/search/movie"
url_Movie = "https://api.themoviedb.org/3/movie/"
api_key = "?api_key=95218016d5bed5424e21ea9e342a9122"

```

```python
# Performing a GET request similar to the one we executed
# earlier
movieNamesDF = pd.read_excel('Film List Filtered to Range.xls')
#movieNamesDF = pd.read_excel('Film List.xls')
resultsDF = pd.DataFrame(columns=['ID', 'TITLE', 'BUDGET', 'REVENUE', 'RUNTIME'])


for index, row in movieNamesDF.iterrows():
    title="&query="+str(row["Title"])
    responseID = requests.get(url_ID + api_key + title)
    data = responseID.json()
    
    pprint(f'Request ID = {responseID.status_code} for movie: {row["Title"]}')
    
    if responseID.status_code == 429:
        print("API LIMIT REACHED")
        time.sleep(10)
        responseID = requests.get(url_ID + api_key +title)
        data = responseID.json()
        
        
    if data["total_results"] > 0:
        responseMovie = requests.get(url_Movie + str(data["results"][0]["id"]) + api_key )
        
        if responseMovie.status_code == 429:
            print("API LIMIT REACHED")
            time.sleep(10)
            responseMovie = requests.get(url_Movie + str(data["results"][0]["id"]) + api_key )
            
        
        dataMovie = responseMovie.json()
        
        genreDict=dataMovie["genres"]
        if len(genreDict)>0:
            genreValue=genreDict[0]["name"]
        else:
            genreValue = ""
            
        if len(genreDict)>1:
            genreValue2=genreDict[1]["name"]
        else:
            genreValue2 = ""

            
        resultsDF = resultsDF.append({'ID':dataMovie["id"], \
                                      'ORIGINAL TITLE': row["Original"],\
                                      'TITLE':dataMovie["title"],\
                                      'BUDGET':dataMovie["budget"],\
                                      'REVENUE':dataMovie["revenue"],\
                                      'RELEASE DATE':dataMovie["release_date"],\
                                      'RUNTIME':dataMovie["runtime"],\
                                      'IMDB_ID':dataMovie["imdb_id"],\
                                      'GENRE 1':genreValue,\
                                      'GENRE 2':genreValue2,\
                                      'POPULARITY':dataMovie["popularity"]},\
                                     ignore_index=True)
        
        

        
    # pprint(f'{responseID.status_code}')
   # pprint(data["total_results"])

    #pprint(data["results"][0]["id"])
    #pprint(data)
    #pprint(data["results"][0]["title"]
  
    #print(response.url)


```

```python
# Converting the response to JSON, and printing the result.

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('results.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
resultsDF.to_excel(writer, sheet_name='Results')

# Close the Pandas Excel writer and output the Excel file.
writer.save()

```
